## What is an Image?
An **image** is a visual representation of something, typically in the form of a 2D matrix of pixel values. Each pixel contains color information (such as RGB values) or intensity (in grayscale). Images can be grayscale, color (e.g., RGB, BGR), or even binary.

- **Grayscale Image:** Each pixel has a single value representing intensity (0 = black, 255 = white).

- **Color Image:** Each pixel has multiple values (e.g., Red, Green, Blue) representing color intensity.

---

## Image Processing and Its Importance
**Image Processing** involves performing operations on images to enhance, analyze, or extract useful information from them. It is widely used in:

- Medical imaging (X-rays, MRIs)

- Object detection and recognition (in self-driving cars, security systems)

- Robotics and automation

- Photography and editing

- Remote sensing and satellite imagery

**Importance:**

- Enhances image quality

- Reduces noise

- Allows for object tracking and feature detection

- Enables machine vision and AI applications

## OpenCV Library

**OpenCV (Open Source Computer Vision Library)** is an open-source library designed for computational efficiency and real-time applications. It provides a vast range of tools for image and video processing.

**Key features:**

- Image filtering, transformations

- Face and object detection

- Video analysis

- Integration with deep learning frameworks

# Operations on an Image
### 1. Load the Image Using OpenCV Library Functions


```python
import cv2
img = cv2.imread('tiger.png', cv2.IMREAD_GRAYSCALE)
```

### 2. Show the Image Using Matplotlib Library

```python
import matplotlib.pyplot as plt


plt.imshow(img, cmap='gray') 
plt.title("Grayscale Image")
plt.axis('off')  
plt.show()
```

### 3. Convert the Image to Other Color Formats and Display

```python
img_color = cv2.imread('tiger.png', cv2.IMREAD_COLOR)

img_rgb = cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB)

img_gray = cv2.cvtColor(img_color, cv2.COLOR_BGR2GRAY)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img_color, cv2.COLOR_BGR2RGB))
plt.title("Original (RGB)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img_color)
plt.title("BGR Image")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(img_gray, cmap='gray')
plt.title("Grayscale Image")
plt.axis('off')

plt.show()
```
