## Introduction to Histogram Equalization

Histogram equalization is an image processing technique used to enhance the contrast of an image by redistributing the pixel intensity values. The idea is to spread out the most frequent intensity values, making the histogram of the image more uniform. This results in images where the contrast is improved, especially in areas that are overly dark or overly bright.

It is commonly used in applications such as medical imaging, satellite images, and photography enhancement to improve the visual quality of images.

---

## Formula and Explanation

Let:

-  r : Original pixel intensity (in range 0â€“255 for 8-bit grayscale)  
-  s: New pixel intensity after equalization  
- $p_r$($r_k$) : Probability of a pixel having intensity $r_k$ 
-  L : Total number of gray levels (usually 256)  
-  n : Total number of pixels in the image  

The histogram equalization transformation function is defined as:

$$
s_k = T(r_k) = (L - 1) \sum_{j=0}^{k} p_r(r_j)
$$

Where:

-  $p_r$($r_k$) = $\frac{n_k}{n}$ is the normalized histogram (PDF)  
-  $\sum_{j=0}^{k} p_r(r_j)$  is the Cumulative Distribution Function (CDF)  
-  $s_k$  is the new intensity corresponding to original level  $r_k$ 
  
---

- **PDF** shows how frequently each intensity level appears.  
- **CDF** gives the cumulative probability up to each intensity level.  
- Multiplying the CDF by  L - 1  maps the values back into the range [0, 255].  

This mapping spreads pixel values across the full dynamic range, enhancing image contrast.


```python
import cv2
import matplotlib.pyplot as plt
```

```python
img = cv2.imread('einstein.png', cv2.IMREAD_GRAYSCALE)
```

```python
equalized = cv2.equalizeHist(img)
```

```python
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
plt.title("Original Image")
plt.imshow(img, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 2)
plt.title("Histogram (Original)")
plt.hist(img.ravel(), 256, color='blue')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.subplot(2, 2, 3)
plt.title("Equalized Image")
plt.imshow(equalized, cmap='gray')
plt.axis('off')

plt.subplot(2, 2, 4)
plt.title("Histogram (Equalized)")
plt.hist(equalized.ravel(), 256, color='green')
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")

plt.tight_layout()
plt.show()
```
