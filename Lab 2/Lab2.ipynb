# LAB 2: Perform Basic Operations on Images

### 1. Image Addition
**Significance of Image Addition in Image Processing:**
- Enhances brightness by adding pixel values.

- Used in image blending (e.g., combining two images).

- Helps in morphing and generating effects like overlay.

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('add_inp1.png', cv2.IMREAD_COLOR)
img2 = cv2.imread('add_inp2.png', cv2.IMREAD_COLOR)

img2_resized = cv2.resize(img2, (img1.shape[1], img1.shape[0]))

added_image = cv2.add(img1, img2_resized)

plt.figure(figsize=(12, 4))

plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img1, cv2.COLOR_BGR2RGB))
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img2_resized, cv2.COLOR_BGR2RGB))
plt.title("Image 2")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(added_image, cv2.COLOR_BGR2RGB))
plt.title("Added Image")
plt.axis('off')

plt.tight_layout()
plt.show()
```

### 2. Image Subtraction
**Significance of Image Subtraction in Image Processing:**
- Highlights differences between images.

- Useful in motion detection, background subtraction, and change detection.

- Can also be used for edge detection when subtracting blurred version from the original.

```python
circle = cv2.imread('sub_inp1.png') 
star = cv2.imread('sub_inp2.png') 
subtracted = cv2.subtract(star, circle) 
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(circle, cv2.COLOR_BGR2RGB))
plt.title("Circle")
plt.axis('off')
plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(star, cv2.COLOR_BGR2RGB))
plt.title("Star")
plt.axis('off')
plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(subtracted, cv2.COLOR_BGR2RGB))
plt.title("Subtracted Image")
plt.axis('off')
```

### 3. Image Multiplication
**Significance of Image Multiplication in Image Processing:**
- Enhances or dims image contrast.

- Used for masking (multiplying with 0s and 1s).

- Can apply gain control in imaging systems.



```python
img5 = cv2.imread('img5.png', cv2.IMREAD_COLOR)
img6 = cv2.imread('img6.png', cv2.IMREAD_COLOR)

img6_resized = cv2.resize(img6, (img5.shape[1], img5.shape[0]))

multiplied_image = cv2.multiply(img5, img6_resized)

plt.figure(figsize=(12, 4))
plt.subplot(1, 3, 1)
plt.imshow(cv2.cvtColor(img5, cv2.COLOR_BGR2RGB))
plt.title("Image 1")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(cv2.cvtColor(img6_resized, cv2.COLOR_BGR2RGB))
plt.title("Image 2")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(cv2.cvtColor(multiplied_image, cv2.COLOR_BGR2RGB))
plt.title("Multiplied Image")
plt.axis('off')
plt.tight_layout()
plt.show()
```

### 4. Image Division
**Significance of Image Division in Image Processing:**
- Used for normalization.

- Helpful in lighting correction, shadow removal.

- In image segmentation to divide overlapping intensity patterns.

```python
img7 = cv2.imread('img7.png', cv2.IMREAD_GRAYSCALE)
img8 = cv2.imread('img8.png', cv2.IMREAD_GRAYSCALE)

img8_resized = cv2.resize(img8, (img7.shape[1], img7.shape[0]))

img7_safe = img7.copy()
img7_safe[img7_safe == 0] = 1

divided_image = cv2.divide(img8_resized, img7_safe)

divided_image_normalized = cv2.normalize(divided_image, None, 0, 255, cv2.NORM_MINMAX)

divided_image_normalized = np.uint8(divided_image_normalized)

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.imshow(img7, cmap='gray')
plt.title("Image 7 (Gray)")
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(img8_resized, cmap='gray')
plt.title("Image 8 (Resized Gray)")
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(divided_image_normalized, cmap='gray')
plt.title("Divided Image (Normalized)")
plt.axis('off')

plt.tight_layout()
plt.show()
```
