## Morphological Image Processing

Morphological operations process images based on their shape. These operations are typically applied to binary or grayscale images for noise removal, object detection, and shape analysis.

### Erosion
Erosion removes pixels from the boundaries of objects, shrinking the foreground.

**Equation**:
$$ (A \circ B)(x, y) = \min\limits_{(u,v) \in B} A(x-u, y-v) $$

### Dilation
Dilation adds pixels to the boundaries of objects, expanding the foreground.

**Equation**:
$$ (A \oplus B)(x, y) = \max\limits_{(u,v) \in B} A(x+u, y+v) $$

### Opening
Opening is an erosion followed by dilation. It removes small objects and noise.

**Equation**:
$$ A \circ B = (A \circ B) \oplus B $$

### Closing
Closing is a dilation followed by erosion. It fills small holes and gaps in objects.

**Equation**:
$$ A \bullet B = (A \oplus B) \circ B $$

These operations help in improving the quality and clarity of images for further analysis.


```python
import numpy as np
import cv2 as cv
from matplotlib import pyplot as plt


img = cv.imread('inpimage.jpg', 0)

kernel = np.ones((5, 5), np.uint8)

erosion = cv.erode(img, kernel, iterations=1)

dilation = cv.dilate(img, kernel, iterations=1)

opening = cv.morphologyEx(img, cv.MORPH_OPEN, kernel)

closing = cv.morphologyEx(img, cv.MORPH_CLOSE, kernel)

plt.figure(figsize=(15, 12))

plt.subplot(3, 2, 1), plt.imshow(img, cmap='gray'), plt.title('Original Image')
plt.subplot(3, 2, 2), plt.imshow(erosion, cmap='gray'), plt.title('Erosion')
plt.subplot(3, 2, 3), plt.imshow(dilation, cmap='gray'), plt.title('Dilation')
plt.subplot(3, 2, 4), plt.imshow(opening, cmap='gray'), plt.title('Opening')
plt.subplot(3, 2, 5), plt.imshow(closing, cmap='gray'), plt.title('Closing')

for i in range(1, 6):
    plt.subplot(3, 2, i).axis('off')

plt.tight_layout()
plt.show()
```
