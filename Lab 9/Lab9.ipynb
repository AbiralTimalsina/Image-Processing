## Haar Cascade Face Detection

### Theory

Haar Cascade Classifier is a machine learning object detection algorithm used to identify objects in images or video. It uses a cascade of simple classifiers to detect faces, eyes, or other objects. Haar Cascade classifiers are trained using positive and negative images, which allow the classifier to learn the patterns of the object (e.g., face) and non-object features.

The algorithm is based on Haar-like features, which are essentially rectangular features that encode the contrast between regions in an image. These features are similar to those used in traditional computer vision techniques like edge detection. The classifier works by scanning the image in a sliding window approach, looking for patterns that match the learned Haar features.

### Key Concepts

1. **Haar-like Features**: These are used to detect edges, lines, and other features in an image. They work by comparing the intensity of adjacent rectangular regions of pixels.

   $$ \text{Haar Feature} = \sum \text{White Rectangle} - \sum \text{Black Rectangle} $$

2. **Integral Image**: This is a technique to compute the sum of pixel values over any rectangular region efficiently. It allows for quick computation of Haar-like features.

   $$ \text{Integral Image at } (x, y) = \sum_{i=0}^{x} \sum_{j=0}^{y} I(i, j) $$

3. **Cascade Classifier**: This is a series of increasingly complex classifiers that process the input image. The classifier works in stages, and the decision at each stage determines whether the image contains the object of interest (e.g., face).

4. **Adaboost**: A machine learning algorithm that combines weak classifiers to form a strong classifier. In the case of Haar Cascades, Adaboost is used to select the best Haar features for detecting faces.

### Applications

- Face detection
- Eye detection
- Pedestrian detection
- Object detection in general

### Haar Cascade Algorithm Steps

1. **Image Preprocessing**: The image is first converted to grayscale, as Haar cascades work on grayscale images for simplicity and speed.
2. **Sliding Window**: The classifier slides a fixed-size window over the image, and at each step, it checks whether the region inside the window matches the learned features.
3. **Face Classification**: The image region is passed through multiple stages of the cascade classifier. If the region passes all stages, it is classified as containing the object (face in this case).
4. **Post-processing**: Once potential faces are detected, further processing like non-maximum suppression can be used to remove duplicate detections.

```python
import cv2
import numpy as np

face_cascade = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')


img = cv2.imread('image.png')

gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

faces = face_cascade.detectMultiScale(gray, 1.1, 5)

for (x, y, w, h) in faces:
    img = cv2.rectangle(img, (x, y), (x + w, y + h), (0, 255, 0), 2)

import matplotlib.pyplot as plt
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.axis('off')
plt.show()

```
